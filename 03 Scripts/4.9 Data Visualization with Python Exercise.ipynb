### 01 Import Libraries and Data

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Create path variable for main project folder
path = r'D:\JupyterProjects\06-2022 Instacart Basket Analysis'
```

```python
# Import data
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'Four8Final.pkl'))
```

### 02 Bar Charts

```python
# Create bar chart
ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
```

```python
# Apply colors to bar chart and create bar object
bar = ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
# Export bar object to Visualizations folder
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

### 03 Histograms and Scatterplots

```python
# Create histogram of price column
ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
# Create scatterplot
sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_merge)
```

```python
# Increase bin for more granularity
hist_2 = ords_prods_merge['prices'].plot.hist(bins = 70)
```

```python
# Export histogram object to Visualizations folder
hist_2.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_prices.png'))
```

### 04 Sampling

```python
# Set random seed
np.random.seed(4)
```

```python
# Create list to store Boolean values
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# Create 70% dataframe
big = ords_prods_merge[dev]
```

```python
# Create 30% dataframe
small = ords_prods_merge[~dev]
```

```python
# Check size
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

### 05 Line Chart of Sampled Data

```python
# Reduce dataframe to only columns needed for chart
df_2 = small[['orders_day_of_week','prices']]
```

```python
df_2.head()
```

```python
# Create line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Export line chart object to Visualizations folder
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line.png'))
```
