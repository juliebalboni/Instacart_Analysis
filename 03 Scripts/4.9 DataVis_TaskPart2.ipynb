### 01 Import Libraries and Data

```python
# Import libraries
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Create path variable for main project folder
path = r'D:\JupyterProjects\06-2022 Instacart Basket Analysis'
```

```python
# Import Data
final = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'merged_final.pkl'))
```

```python
# Check dataframe dimensions
final.shape
```

```python
# Reduce dataframe to columns needed for exercise due to memory contraints
final = final.filter(['orders_day_of_week', 'order_hour_of_day', 'prices', 'loyalty_flag','age', 'no_of_dependents', 'income'])
```

```python
# Check first five rows
final.head()
```

### 02 Create Bar Chart for Frequency of Orders by Day of Week

```python
 bar_color = final['orders_day_of_week'].value_counts().sort_index().plot.bar(color = ['red', 'pink', 'orange', 'yellow', 'green', 'blue', 'purple'])
```

```python
# Export color bar chart
bar_color.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_color.png'))
```

### 03 Create Histogram of Prices

```python
# Histogram
histogram = final['prices'].plot.hist(bins=70)
```

```python
# Export histogram
histogram.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'histogram.png'))
```

### 04 Create Scatterplot of Prices

```python
# Scatterplot
scatterplot = sns.scatterplot(x = 'prices', y = 'prices',data = final)
```

```python
# Export scatterplot
scatterplot.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'scatter_price.png'))
```

### 05 Create Line Chart Using Sample Subset

```python
# Set random seed
np.random.seed(4)
```

```python
# Create list to store Boolean values
dev = np.random.rand(len(final)) <= 0.7
```

```python
# Create 70% dataframe
big = final[dev]
```

```python
# Create 30% dataframe
small = final[~dev]
```

```python
# Check size
len(final)
```

```python
len(big) + len(small)
```

```python
# Reduce dataframe to only columns needed for chart
df_2 = small[['orders_day_of_week','prices']]
```

```python
df_2.head()
```

```python
# Line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
```

### 06 Create Histogram of Orders by Hour of Day

```python
hist_ohd = final['order_hour_of_day'].plot.hist(bins=70)
```

```python
# Export histogram
hist_ohd.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_ohd.png'))
```

The histogram shows that most orders are placed between the 9th and 17th hours of the day with a slight peak at the 10th hour.

### 07 Create Loyalty Flag Bar Chart

```python
bar_loyalty = final['loyalty_flag'].value_counts().sort_index().plot.bar(color = ['blue', 'green', 'purple'])
```

```python
# Export bar chart
bar_loyalty.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_loyalty.png'))
```

### 08 Create Line Chart of Difference in Expenditure Depending on the Hour of the Day

```python
# Reduce small dataframe to only columns needed for chart
df_3 = small[['order_hour_of_day','prices']]
```

```python
line_ohd_price = sns.lineplot(data = df_3, x = 'order_hour_of_day', y = 'prices')
```

```python
# Export line chart
line_ohd_price.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_ohd_price.png'))
```

### 09 Create Line Chart of Age and Number of Dependents

```python
# Reduce small dataframe to only columns needed for chart
df_4 = small[['no_of_dependents','age']]
```

```python
line_age_dep = sns.lineplot(data = df_4, x = 'age', y = 'no_of_dependents')
```

There does not appear to be a strong relationship between age and number of dependents.

```python
# Export line chart
line_age_dep.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_age_dep.png'))
```

### 10 Create Scatterplot of Age and Spending Power

```python
# Scatterplot
scatter_age_income = sns.scatterplot(x = 'age', y = 'income', data = final)
```

Most customers make $200,000 or less.  There is a distinct jump in income for customers 40 and older giving older customers the most spending power.

```python
# Export scatterplot
scatter_age_income.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'scatter_age_income.png'))
```
